```python
# Фнукция для подсчета заполненности
def data_fills_count(df):
    return_lst = ["Доля пустых значений:"]
    return_lst.append(df.isna().sum() / df.shape[0])
    proc = 1 - (df.isna().sum().sum() / df.size)
    return_lst.append("Заполненность всей таблицы данными: {:.1%}".format(proc))
    return return_lst

data_fills_count(df)
```

В предоставленных данных могут быть пустые и дублированные записи.
Попробуем их выявить и удалить дублированные и полностью пустые строки/столбцы.
**Функция возвращает `dataframe` с преобразованными типами данных.**

```python
# Создадим функцию удаления строк и столбцов полностью состоящих из пустых значений или дубликатов
def clean_clear(df):
    print("Попытка удаления полностью пустых или дубликатов столбцов и строк...")
    # Сохраняем размерность датафрейма
    shape_sum = df.shape

    # Удаляем дубликаты строк
    df = df.drop_duplicates()
    # Удаляем дубликаты столбцов
    df = df.T.drop_duplicates().T

    # Удаляем полностью пустые столбцы и строки
    df = df.dropna(axis=0, how='all')
    df = df.dropna(axis=1, how='all')

    # Выводим результаты
    print(f"Форма датасета до операции: {shape_sum}")
    print(f"Форма датасета после операции: {df.shape}")

    print(f"Количество строк после удаления пустых и повторяющихся: {df.shape[0]}")
    print(f"Количество столбцов после удаления пустых и повторяющихся: {df.shape[1]}")
    print(f"***Удалено {shape_sum[0] - df.shape[0]} строк и {shape_sum[1] - df.shape[1]} столбцов.")
    df = df.convert_dtypes()
    return df

df = clean_clear(df)
print("Типы данных:")
display(df.dtypes.value_counts())
```

Параметр `include=['all']` метода `.describe()` выведет статистику для всех данных:

- **unique** - количество уникальных элементов
- **top** - самый распространенный элемент
- **freq** - сколько раз встречается самый распространенный элемент

Найдем столбцы с GUID (`ДокументПрохождениеМедосмотра`, `Клиент`, `Контрагент`, etc.) с помощью модуля `regex` выражения и удалим их (через замену на `np.nan` значения).
Найдем значения `'<NA>'`, которые преобразовались в строки, заменим их на `np.nan` и по возможности удалим пустые строки/столобцы.

```python
# Функция для поиска столбцов с GUID и удаление их
def is_GUID_or_NA(df, to_replace, value):
    df = df.replace(to_replace=to_replace, value=np.nan, regex=True)
    print(f"Колонки для удаления:\n{df.columns[df.isna().all()].tolist()}")
    df = clean_clear(df)
    return df

# Найти и удалить столбцы с GUID
regex = "^[{]?[0-9a-fA-F]{8}" + "-([0-9a-fA-F]{4}-)" + "{3}[0-9a-fA-F]{12}[}]?$"
df = is_GUID_or_NA(df, regex, np.nan)

# Найти и заменить '<NA>'
df = is_GUID_or_NA(df, '<NA>', np.nan)
df.head()
```

Еще раз почистим полностью пустые/дублированные строки и столбцы, если таковые появились и посмотрим как изменилась доля заполненных значений.
