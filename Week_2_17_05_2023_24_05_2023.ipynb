Вообщем и целом видим какое-то количество пропусков есть.

```python
#Доля пропусков от общего количества элементов
data.isna().sum() / data.shape[0]
```

```python
# Общий процент пропусков
proc = data.isna().sum().sum() / data.size
print('{:.1%}'.format(proc))
```
