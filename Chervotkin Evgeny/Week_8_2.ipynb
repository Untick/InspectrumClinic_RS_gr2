**Удалим пустые столбцы**

```python
# Удаление пустых столбцов
data_columns = data.dropna(axis=1, how='all')

# Вывод обновленного датафрейма
data_columns
```

**Было 289 столбцов, стало 214**

Удалим полностью дублирующие столбцы.

```python
# Удаление полностью дублирующихся столбцов
data_duplicate = data_columns.T.drop_duplicates().T

# Вывод обновленного датафрейма
data_duplicate
```

**Дубликаты удалены, стало 150 столбцов вместо 214**

Теперь избавимся от GUIDов.

```python
import re
import pandas as pd


df = pd.DataFrame(data_duplicate)

# Функция для удаления столбцов с GUID
def remove_guid_columns(df):
    # Паттерн для сопоставления GUID
    regex_GUID = r"\b[A-Fa-f0-9]{8}(?:-[A-Fa-f0-9]{4}){3}-[A-Fa-f0-9]{12}\b"

    # Поиск и удаление столбцов с GUID
    columns_to_remove = []
    for column in df.columns:
        if re.search(regex_GUID, str(df[column])):
            columns_to_remove.append(column)

    df = df.drop(columns=columns_to_remove)
    return df

# Удаление столбцов с GUID
df_guid = remove_guid_columns(df)
```

```python
df_guid
```

Всего 3 колонки с GUID.

**Теперь обработаем колонку ЗаключениеМК**

```python
# Подсчет количества значений с учетом пустых значений
value_counts = df_guid['ЗаключениеМК'].value_counts(dropna=False)

# Вывод результатов
print(value_counts)
```

```python
# Получение уникальных значений
unique_values = df_guid['ЗаключениеМК'].unique()

# Вывод результатов
print(unique_values)
```

```python
# Удаление пробелов в начале и конце строк
df_guid['ЗаключениеМК'] = df_guid['ЗаключениеМК'].str.strip()
```

Удалил пробелы в названиях, теперь уберу мусорные строки. Предварительно преобразовав данные в тип данных 'str' иначе не сработает.

```python
# Преобразование столбца в тип данных 'str'
df_guid['ЗаключениеМК'] = df_guid['ЗаключениеМК'].astype(str)
```

```python
# Создание маски для удаления строк с определенными значениями
mask = ~df_guid['ЗаключениеМК'].str.contains('Газоспасатель|п. 11.3|машинист бульдозера')

# Фильтрация датафрейма и удаление строк
df_cleaned = df_guid[mask]

# Вывод обновленного датафрейма
df_cleaned
```

Проверим колонку ЗаключениеМК

```python
# Подсчет количества значений с учетом пустых значений
value_counts = df_cleaned['ЗаключениеМК'].value_counts(dropna=False)

# Вывод результатов
print(value_counts)
```

**Теперь приведем данные к 4 значениям: Годен, ОграниченноГоден, Негоден и nan. Оставляю Nan так как из него хочу вынуть значения и добавить в негоден.**

```python
# Замена значений в столбце 'ЗаключениеМК'
df_cleaned['ЗаключениеМК'] = df_cleaned['ЗаключениеМК'].replace({
    'ГоденСКоррекциейЗрения': 'ОграниченноГоден',
    'ГоденБезРаботНаВысотах': 'ОграниченноГоден',
    'ГоденСКоррекциейСлуха': 'ОграниченноГоден',
    'НуждаетсяВДообследованииИЛечении': 'Негоден'
})

# Вывод обновленного датафрейма
df_cleaned
```

```python
# Подсчет количества значений с учетом пустых значений
value_counts = df_cleaned['ЗаключениеМК'].value_counts(dropna=False)

# Вывод результатов
print(value_counts)
```

**Выделю строки содержащие в столбце значения nan в отделный датафрейм для дальнейшего анализа.**
