```python
# Удаление столбцов которые явно не нужны
df_very_cleaned = df_very_cleaned.drop(['ЗаключениеМКПредставление', 'Заключение','ДатаЗавершенияМедосмотра','ТипМедосмотра'], axis=1)
```

Нужно выяснить какие типы данных в столбцах, чтобы грамотно их перевести в цифры.

```python
# Проход по каждому столбцу таблицы
for column in df_very_cleaned.columns:
    # Получение уникальных значений в столбце, включая пустые ячейки
    unique_values = df_very_cleaned[column].unique()

    # Подсчет количества значений, включая пустые ячейки
    value_counts = df_very_cleaned[column].value_counts(dropna=False)

    # Вывод результатов с выделением названия столбца
    print(f"******************* Столбец: {column} *******************") #Выделим звездочками название Столбец
    print(f"Уникальные значения: {unique_values}")
    print(f"Количество значений: \n{value_counts}\n")
```

Поменялось местами ЗаключениеМК. Повернём обратно.

```python
# Замена значений в столбце 'ЗаключениеМК'
df_cleaned['ЗаключениеМК'] = df_cleaned['ЗаключениеМК'].replace({
    'Годен': 'Негоден',
    'Негоден': 'Годен'
})
```

Столбец Цех и столбец Профессия нужно обрабатывать Bag of Words.

**Удалило очень много данных** больше половины столбцов, что может быть ошибка. Поэтому создал новую переменную. Ошибка может быть в том, что данные эти перекрывались другими данными. Например если в этих столбцах нет данных, то они скорее всего вытекали из других столбцов, а мы эти данные безжалостно отрубили.

Преобразуем столбец датаРождения в числовой формат.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Создание копии датафрейма с преобразованным столбцом
df_normalized = df_very_cleaned.copy()

# Преобразование столбца в числовой формат
df_normalized['КлиентГодРождения'] = df_normalized['КлиентГодРождения'].astype(int)

# Нормализация значений столбца
scaler = MinMaxScaler()
df_normalized['КлиентГодРождения'] = scaler.fit_transform(df_normalized['КлиентГодРождения'].values.reshape(-1, 1))

# Вывод обновленного датафрейма
df_normalized
```

```python
# Удаление несоответствующих колонок
df_normalized = df_normalized[(df_normalized['КлиентПол'] == 'Женский') | (df_normalized['КлиентПол'] == 'Мужской')]
```

```python
# Присвоим значение 1 к мужскому полу и 0 к женскому.
df_normalized['КлиентПол'] = df_normalized['КлиентПол'].replace({'Мужской': 1, 'Женский': 0})
```

```python
import numpy as np

# Переименование значений на цифровой формат
df_normalized['ПсихОсвидетельствование'] = df_normalized['ПсихОсвидетельствование'].replace({True: 1, False: 0, np.nan: 2})

# Удаление ненужных значений
df_normalized = df_normalized.dropna(subset=['ПсихОсвидетельствование'])
```

```python
# Присваеваем число каждому слову в колонке цех
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_normalized['Цех'] = label_encoder.fit_transform(df_normalized['Цех'])
```

```python
# Присваеваем число каждому слову в колонке профессия
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_normalized['Профессия'] = label_encoder.fit_transform(df_normalized['Профессия'])
```

```python
# Удаление кириллицы из колонки вредные факторы.
import re
# Определение паттерна для поиска кириллических символов
pattern = re.compile('[а-яА-Я]')

# Фильтрация и удаление строк с кириллическими записями
df_filtered = df_normalized[~df_normalized['ВредныеФакторы'].astype(str).str.contains(pattern)]
```

Сохраним обработанный датасет.

```python
df_filtered.to_csv('путь_к_файлу.csv', index=False)
```
